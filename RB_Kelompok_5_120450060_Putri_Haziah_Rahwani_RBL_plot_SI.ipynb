# Plot SI

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## **fitness individu saat iterasi terakhir**

# 25 Generasi

```python
dat=pd.read_csv('hasil_0.csv')
d=dat.rename(columns={'14':'jarak'})
```

```python
d2=pd.DataFrame(d['jarak'])
d2['Iterasi'] = range (d2.shape[0])
```

```python
plt.figure(figsize=(10,10))
ax=plt.gca()
ax.set_title('Semut Terbaik per Iterasi', fontsize=25)
ax.set_ylabel('Jarak (m)', fontsize = 18)
ax.set_xlabel('individu ant', fontsize = 18)
d2.plot(ax=ax, kind='line', y='jarak', x='Iterasi')
ax.grid()
```

**solusi TSP dalam populasi ACO**

**Rute terbaik**

```python
dat.iloc[0,:-1].astype(int)
```
